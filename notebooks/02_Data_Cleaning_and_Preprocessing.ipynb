```python
missing=x_train.isnull().mean()
missing
```

```python
missing.sort_values(ascending=False)
```

```python
#GET THE COLUMNS WITH MORE THAN 60% MISSING VALUES
more_than_60=missing[missing>0.6].index.tolist()
```

```python
#CREATE FLAGS FOR THOSE COLS
for col in more_than_60:
    x_train[col+"_missing_flag"]=x_train[col].isnull().astype(int)
    x_test[col+"missing_flag"]=x_test[col].isnull().astype(int)
```

```python
len(more_than_60)
```

```python
#DO IMPUTATION FOR THOSE COLS
def imputation_for_missing_values(df1,df2,cols):
    for col in cols:
        if df1[col].dtype=='object':
            df1.loc[(df1[col].isnull()),col]='Missing'
            df2.loc[(df2[col].isnull()),col]='Missing'
        else:
            df1.loc[(df1[col].isnull()),col]=df1[col].median()
            df2.loc[(df2[col].isnull()),col]=df1[col].median()
```

```python
all_missing_cols=missing[missing>0].index.tolist()
```

```python
imputation_for_missing_values(x_train,x_test,all_missing_cols)
```
